# Get all battles within from specified clan

```python
def getBattles(clan, apikey):
    '''Returns battlelog from a designated clan
    
    Args: 
        Clan: string for clan's tag
        apikey: string for apikey
    Returns:
        battlelog: json battle log
    '''
    headers = {"Authorization": "Bearer "+apikey}
    battlelog = requests.get("https://api.royaleapi.com/clan/"+clan+"/battles", headers=headers)
    return battlelog.json()
```

# Clean Battle Log Data

```python
def cleanBattleLog(battlelog):
    '''Clean up a data frame filled with battlelog data.
    Args: 
        battlelog: JSON containing CR battlelog info
    Returns:
        battleDF: DF of useful information from battlelog'''
    #filter for clanMate battles (autofiltered)
    cols = ['deck', 'crownsEarned', 'name', 'tag', 'utcTime']
    oppDecks = json_normalize(battlelog, ['opponent'], 'utcTime')[cols]
    teamDecks = json_normalize(battlelog, ['team'], 'utcTime')[cols]
    battleDF = oppDecks.join(teamDecks, lsuffix='_opp', rsuffix='_team')
    #drop duplicates created by spectators (creates multiple of the same event)
    battleDF['id'] = battleDF.apply(lambda x: str(sorted([x['name_opp'], x['name_team'], str(x['crownsEarned_opp']), str(x['crownsEarned_team'])])), axis =1)
    battleDF = battleDF.drop_duplicates(['id', 'utcTime_team'])
    battleDF['winner'] =  battleDF.apply(lambda x: x['name_team'] if x['crownsEarned_team'] > x['crownsEarned_opp'] else x['name_opp'], axis=1)
    battleDF['winnerTag'] =  battleDF.apply(lambda x: x['tag_team'] if x['crownsEarned_team'] > x['crownsEarned_opp'] else x['tag_opp'], axis=1)
    battleDF['deck_opp'] = battleDF.apply(lambda i: sorted([x['name'] for x in i['deck_opp']]),axis=1)
    battleDF['deck_team'] = battleDF.apply(lambda i: sorted([x['name'] for x in i['deck_team']]),axis=1)
    battleDF['utcTime'] = battleDF['utcTime_opp']
    battleDF = battleDF.drop(['utcTime_opp', 'utcTime_team'], axis=1)
    return battleDF

def saveBattleDF(clanName, battleDF):
    '''Saves Battlelog data to csv in the form of a Pandas DataFrame
    Args:
        clanName: name of Clan
        battleDF: battle data cleaned and in a DataFrame
    Returns:
        None
    '''
    #toFinish
```
